# Task : Analyse Youtube data of Mr.Beast and help improve his strategy and grow his audience

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("mrbeast_youtube_data.csv")
```

```python
df.head()
```

```python
df.columns
```

## 1) Views Over Time

```python
plt.figure(figsize=(14, 6))
plt.plot(df['published_at'], df['views'], marker='o')
plt.title("1. Views Over Time")
plt.xlabel("Publish Date")
plt.ylabel("Views")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2) Top 10 Most Viewed Videos

```python
top_views = df.sort_values(by='views', ascending=False).head(10)
sns.barplot(data=top_views, x='views', y='title', palette='mako')
plt.title("2. Top 10 Most Viewed Videos")
plt.xlabel("Views")
plt.ylabel("Title")
plt.tight_layout()
plt.show()
```

## 3)  Top 10 Most Liked Videos

```python
top_likes = df.sort_values(by='likes', ascending=False).head(10)
sns.barplot(data=top_likes, x='likes', y='title', palette='viridis')
plt.title("3. Top 10 Most Liked Videos")
plt.xlabel("Likes")
plt.ylabel("Title")
plt.tight_layout()
plt.show()
```

## 4) Top 10 Most Commented Videos

```python
top_comments = df.sort_values(by='comments', ascending=False).head(10)
sns.barplot(data=top_comments, x='comments', y='title', palette='plasma')
plt.title("4. Top 10 Most Commented Videos")
plt.xlabel("Comments")
plt.ylabel("Title")
plt.tight_layout()
plt.show()
```

## 5) Likes vs Views

```python
sns.scatterplot(data=df, x='views', y='likes')
plt.title("5. Likes vs Views")
plt.xlabel("Views")
plt.ylabel("Likes")
plt.tight_layout()
plt.show()
```

## 6) Comments vs Views

```python
sns.scatterplot(data=df, x='views', y='comments')
plt.title("6. Comments vs Views")
plt.xlabel("Views")
plt.ylabel("Comments")
plt.tight_layout()
plt.show()
```

## 7) Likes vs Comments

```python
sns.scatterplot(data=df, x='likes', y='comments')
plt.title("7. Likes vs Comments")
plt.xlabel("Likes")
plt.ylabel("Comments")
plt.tight_layout()
plt.show()
```

## 8) Views Distribution

```python
plt.hist(df['views'], bins=30, color='skyblue', edgecolor='black')
plt.title("8. Distribution of Views")
plt.xlabel("Views")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## 9) Word Cloud of Titles (Winnning key words)

```python
from wordcloud import WordCloud

text = " ".join(df['title'].tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("9. Word Cloud of Video Titles")
plt.tight_layout()
plt.show()
```

## 10) Correlation Heatmap

```python
corr = df[['views', 'likes', 'comments']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("10. Correlation Heatmap")
plt.tight_layout()
plt.show()
```
